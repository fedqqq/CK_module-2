```python
fig, ax = plt.subplots(4)

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.sin(x))
ax[2].plot(x, np.sin(x))
ax[3].plot(x, np.sin(x))

ax[1].set_xlim(-2, 12)
ax[1].set_ylim(-1.5, 1.5)

ax[2].set_xlim(12, -2)
ax[2].set_ylim(1.5, -1.5)

ax[3].autoscale(tight=True)

plt.show()
```

```python
plt.subplot(3, 1, 1)
plt.plot(x, np.sin(x))
plt.title("Cинус") #название
plt.xlabel('x')
plt.ylabel('sin(x)')

plt.subplot(3, 1, 2)
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), '-.y', label='cos(x)')
plt.title("Cинус и косинус") #название
plt.xlabel('x')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), '-.y', label='cos(x)')
plt.title("Cинус и косинус") #название
plt.xlabel('x')
plt.axis('equal')
plt.legend()


plt.subplots_adjust(hspace=1.5)
plt.show()
```

```python
rng = np.random.default_rng(0)

colors = rng.random(30)
sizes = 30 * rng.random(30)

plt.scatter(x, np.sin(x), marker='o', c=colors, s=sizes)
plt.colorbar()
plt.show()
```

если точек больше 1000, то plot предпочтительнее из-за производительности

```python
x = np.linspace(0, 10, 50)

dy = 0.4
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k')
plt.fill_between(x, y - dy, y + dy, color='r', alpha=0.4)
plt.show()
```

```python
def f(x, y):
    return np.sin(x) ** 5 + np.cos(20 + x * y) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

#plt.contour(X, Y, Z, cmap='RdGy')
#plt.contourf(X, Y, Z, cmap='RdGy')
c = plt.contour(X, Y, Z)
plt.clabel(c)
plt.imshow(Z, extent=[0, 5, 0, 5], cmap='RdGy', interpolation='gaussian',
           origin='lower', aspect='equal')
plt.colorbar()
plt.show()
```
