# Гистограммы

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Гистограмма
rng = np.random.default_rng(1)
data = rng.normal(size=1000)
plt.hist(data)
plt.show()
```

```python
plt.hist(data,
         bins=30,
         density=True,
         alpha=0.5,
         histtype='step',
         edgecolor='red',
        )
plt.show()
```

```python
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)

args = dict(
    alpha=0.3,
    bins=40
)

plt.hist(x1, **args)
plt.hist(x2, **args)
plt.hist(x3, **args)
plt.show()
```

```python
print(np.histogram(x1, bins=1))
print(np.histogram(x1, bins=2))
print(np.histogram(x1, bins=40))
```

```python
#Двумерные гистограммы
mean = [0, 0]
cov = [[1, 1], [1, 2]]

x, y = rng.multivariate_normal(mean, cov, 10000).T

#plt.hist2d(x, y, bins=30)

plt.hexbin(x, y, gridsize=30)
cb = plt.colorbar()
cb.set_label('Point in interval')
plt.show()
```

```python
print(np.histogram2d(x, y, bins=1))
print(np.histogram2d(x, y, bins=10))
```

```python
x = np.linspace(0, 10, 1000)

y = np.sin(x[:,np.newaxis] + np.pi * np.arange(0, 2, 0.5))

lines = plt.plot(x, y)
plt.axis('equal')

plt.legend(lines, ['1', 'второй', 'third', '4-ый'], loc='center')
plt.show()
```

```python
import pandas as pd
cities = pd.read_csv('california_cities.csv')
cities.head()
```

```python
lat, lon, pop, area = cities['latd'], cities['longd'], cities['population_total'], cities['area_total_sq_mi']
plt.scatter(lon, lat, c=np.log10(pop), s=area)

plt.xlabel('Широта')
plt.ylabel('Долгота')

cd = plt.colorbar()
cd.set_label('Население')

plt.scatter([], [], c='k', alpha=0.5, s=100, label='100 $km^2$')
plt.scatter([], [], c='k', alpha=0.5, s=300, label='100 $km^2$')
plt.scatter([], [], c='k', alpha=0.5, s=500, label='100 $km^2$')

plt.clim(3, 7)
plt.legend(labelspacing=3, frameon=False)
plt.show()
```

```python
import matplotlib as mpl
fig, ax = plt.subplots()

lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0, 10, 1000)
for i in range(4):
    lines += ax.plot(
        x, 
        np.sin(x - i + np.pi / 2),
        styles[i]
    )

ax.axis('equal')
ax.legend(lines[:2], ['line1', 'line2'], loc='upper right')
leg = mpl.legend(ax, lines[1:], ['line2', 'line3', 'line4'], loc='lower left')
ax.add_artist(leg)
plt.show()
```
