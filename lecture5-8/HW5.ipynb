```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(8, 5))

X = np.array([1.0, 5.0, 10.0, 15.0, 20.0])
red_Y = np.array([0.5, 7.0, 3.0, 5.0, 11.0])
green_Y = np.array([4.0, 3.0, 0.5, 8.0, 12.0])

plt.plot(X, red_Y, '-or', label='line1')
plt.plot(X, green_Y, '-.og', label='line1')
plt.legend()
```

```python
grid = plt.GridSpec(2, 2)

plt.subplot(grid[0, :2])
plt.plot(np.arange(1, 6), np.array([1, 7, 6, 3, 4]))

plt.subplot(grid[1, 0])
plt.plot(np.arange(1, 6), np.array([9, 4, 2, 4, 9]))

plt.subplot(grid[1, 1])
plt.plot(np.arange(1, 6), np.array([-7, -4, 2, -4, -7]))

plt.show()
```

```python
X = np.arange(-5, 6)
Y = np.array([i ** 2 for i in X])

plt.plot(X, Y)

plt.annotate('min', xy=(0, 0), xytext=(0, 10), arrowprops=dict(facecolor='green'))
plt.show()
```

```python
np.random.seed(0)
data = np.random.randint(0, 11, size=(8, 8))

matrix = data.reshape((8, 8))

fig, ax = plt.subplots()
im = ax.imshow(matrix, cmap='viridis', interpolation='none')

cb = fig.colorbar(im, shrink=0.5)
cb.ax.set_ylabel(rotation=-90, va="bottom")

plt.show()
```

№5

```python
x = np.linspace(0, np.pi * 5, 100)
y = np.sin(x)

plt.plot(x, y, 'r')
plt.fill_between(x, y, facecolor='blue', interpolate=True)

plt.show()
```

```python
X = np.linspace(0, np.pi * 5, 100)
y = np.array([np.cos(x) for x in X if np.cos(x) > -0.5])

plt.plot(np.array([x for x in X if np.cos(x) > -0.5]), y)


plt.show()
```

```python
y = np.array([10, 15, 25, 15, 35])
mylabels = ["Toyota", "Ford", "Jaguar", "Audi", "BMW"]
myexplode = [0, 0, 0, 0, 0.2]

plt.pie(y, labels = mylabels, explode = myexplode, startangle = -150)
plt.show()
```
